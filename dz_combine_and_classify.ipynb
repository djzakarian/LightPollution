### 2) Classify Data
Loop through the data tables associated with my_locs and classify nights and rows as clear, overcast, or neither

CLEAR: 

    - Confidence = True: 0 okta, no moon, no sun
    - Confidence = False: 0-2 okta, no moon, no sun

OVERCAST: 

    - Confidence = True: 8 okta, no moon, no sun
    - Confidence = False: 6-8 okta, no moon, no sun

```python
my_locs = ['Rolla', 'MAC', 'SEMO', 'Timberlane', 'Ozark', 'Ozark_II', 
           'Perryville', 'Overland', 'Hillsboro', 'Science Center', 
           'Magruder Hall', 'MG Hall', 'TSO', 'Thousand_Hills', 'Grand Gulf', 
           'echo bluff', 'Jefferson Arch', 'Broemmelsiek Park']
# my_locs =['Broemmelsiek Park']
```

```python
""" 
Read in the data tables and update them with the classifications
"""

# look through the files in the directory 
# and pick out the data tables associated with my_locs

# my_locs = ['TSO']


directory = "Combined Data Tables"
files = os.listdir(directory)

for location in my_locs:  
  
    # Filter only CSV files containing the desired location in their filename
    location = location.replace(' ', '_')
    csv_filenames = [file for file in files if file.endswith('.csv') and location in file]

# classify and update tables 

    for df_filename in csv_filenames:
        df_path = os.path.join("Combined Data Tables", df_filename)
        df = pd.read_csv(df_path, low_memory=False)
        df = classify_rows(df)
        df = classify_nights(df)
        # Save the updated DataFrame to the same path
        df.to_csv(df_path, index=False)  # Setting index=False to avoid saving the DataFrame index
```
